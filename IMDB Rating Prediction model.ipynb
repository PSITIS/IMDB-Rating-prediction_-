# ML model

Importing Libraries

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

## Selecting relevant deatures for Model

```python
df = new_record_2.drop(columns=['released','imdbid'], axis =1)

#converting genre list values to  string
df['genre'] = df['genre'].apply(str).str.replace("[",'')
df['genre'] = df['genre'].apply(str).str.replace("]",'')
df['genre'] = df['genre'].apply(str).str.replace("'",'')

df.head()
```

# encoding categorical value for regression model

```python

encoding_df = pd.get_dummies(df,drop_first=True,columns=['genre','title','synopsis'])
```

```python
X = encoding_df.drop('imdbrating',axis =1)
y = encoding_df['imdbrating']
```

```python
X.head()
```

```python
y.head()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=30)
```

```python
LR = LinearRegression()
LR.fit(X_train,y_train)
```

```python
y_pred = LR.predict(X_test)
```

# Accuracy of Model

```python
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
missing_cols = set(encoding_df.columns) - set(sample_df_encoding.columns)
for col in missing_cols:
    sample_df_encoding[col] = 0
```

# To reorder sample df columns as training data contained

```python

sample_df_encoding = sample_df_encoding[encoding_df.columns]
```

 #  I have droped the IMDB rating column in sample encoded df so that it fits as an input to our model with no extra parameters
